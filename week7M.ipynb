```python
import matplotlib
import matplotlib.pyplot as a
import numpy as b
```

```python
fig = a.figure()
ax = a.axes(projection = '3d')
a.show()
#memunculkan grafik 3D tanpa data
```

```python
z = b.linspace(0, 1, 100)
x = z * b.sin(20 * z)
y = z * b.cos(20 * z)
#data untuk 3d plot
```

```python
fig = a.figure()
ax = a.axes(projection = '3d')
ax.plot(x, y, z, color = 'red')
ax.set(title = 'contoh 3D line plot',
       xlabel = 'sumbu x',
       ylabel = 'sumbu y',
       zlabel = 'sumbu x')
a.show()
#membuat 3d line plot dengan data diatas
```

```python
z = b.linspace(0, 1, 100)
x = z * b.sin(20 * z)
y = z * b.cos(20 * z)
#data untuk 3d plot
```

```python
fig = a.figure()
ax = a.axes(projection = '3d')
ax.scatter(x, y, z, color = 'green')
ax.set(title = 'contoh 3D scatter plot',
       xlabel = 'sumbu x',
       ylabel = 'sumbu y',
       zlabel = 'sumbu x')
a.show()
#membuat 3d scatter plot dengan data diatas
```

```python
n = b.linspace(-3.0, 3.0, 100)
x, y = b.meshgrid(n, n)
z = b.sqrt(x**2 + y**2)
print(x.shape)
print(y.shape)
print(z.shape)
#Dataset
```

```python
a.contour(x, y, z)
a.colorbar()
a.title('countour plot')
a.xlabel('sumbu x')
a.ylabel('sumbu y')
a.show()
#membuat contour plot dengan dataset diatas
```

```python
a.contourf(x, y, z)
a.colorbar()
a.title('filled countour plot')
a.xlabel('sumbu x')
a.ylabel('sumbu y')
a.show()
#membuat filled contour plot dengan dataset diatas
```

```python
n = b.linspace(-3.0, 3.0, 100)
x, y = b.meshgrid(n, n)
z = b.sqrt(x**2 + y**2)
print(x.shape)
print(y.shape)
print(z.shape)
```

```python
fig = a.figure()
ax = a.axes(projection = '3d')
ax.plot_wireframe(x, y, z, color = 'orange')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D wireframe')
a.show()
#membuat grafik 3d wireframe
```

```python
fig = a.figure()
ax = a.axes(projection = '3d')
#ax.plot_surface(x, y, z, color = 'orange')
ax.plot_surface(x, y, z, cmap = 'viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D surface plot')
a.show()
#membuat 3d surface plot
```
