```python
import matplotlib
import matplotlib.pyplot as a
import pandas as b
#import module dan variabel
```

```python
df = b.DataFrame({
    'nama' : ['cecep', 'karti', 'bejo', 'tejo', 'asep', 'wati', 'amat'],
    'usia' : [23, 78, 22, 19, 45, 33, 20],
    'gender' : ['L', 'P', 'L', 'L', 'L', 'P', 'L'],
    'provinsi' : ['jabar', 'jatim', 'jabar', 'jatim', 'jabar', 'jateng', 'jateng'],
    'n_anak' : [2, 0, 0, 3, 2, 1, 4],
    'n_pets' : [5, 1, 0, 5, 2, 2, 3]
})
df
#buat dataset
```

```python
df.plot(kind = 'scatter', x = 'n_anak', y = 'n_pets', color = 'red')
a.title('Scatter Plot')
a.xlabel('jumlah anak')
a.ylabel('jumlah pets')
a.show()
#membuat grafik scatterplot dengan dataset diatas
```

```python
df[['usia']].plot(kind = 'hist', bins = [0, 20, 40, 60, 80, 100])
a.title('Histogram')
a.xlabel('usia')
a.ylabel('jumlah')
a.show()
#membuat grafik histogram dengan dataset diatas
```

```python
df.plot(kind = 'line', x = 'nama', y = 'n_anak', color = 'green')
a.title('line plot')
a.ylabel('jumlah')
a.show()
#membuat grafik lineplot dengan dataset diatas
```

```python
df.plot(kind = 'line', x = 'nama', y = 'n_anak', color = 'green')
df.plot(kind = 'line', x = 'nama', y = 'n_anak', color = 'red')
a.title('line plot')
a.ylabel('jumlah')
a.show()
#membuat grafik multiple lineplot dengan dataset diatas
```

```python
ax = a.gca()
df.plot(kind = 'line', x = 'nama', y = 'n_anak', color = 'green', ax = ax)
df.plot(kind = 'line', x = 'nama', y = 'n_anak', color = 'red', ax = ax)
a.title('line plot')
a.ylabel('jumlah')
a.show()
#sama tapi disatuin
```

```python
df.plot(kind = 'bar', x = 'nama', y = 'usia')
a.title('bar plot')
a.ylabel('jumlah')
a.show()
#membuat grafik bar plot dengan dataset diatas
```

```python
df.groupby('provinsi')['nama'].count().plot(kind = 'bar')
a.title('Bar plot')
a.ylabel('jumlah')
a.show()
#membuat grafik bar plot dengan pengelompokan provinsi
```

```python
df.groupby(['provinsi', 'gender']).size().unstack().plot(kind = 'bar',
                                                         stacked = True)
a.title('stacked Bar plot')
a.ylabel('jumlah')
a.show()
#buat grafik stacked barplot. isinya data laki laki dan perempuan yang ditumpuk dalem 1 bar dan dibedakan
```

```python
df.groupby(['gender', 'provinsi']).size().unstack().plot(kind = 'bar',
                                                         stacked = True)
a.title('stacked Bar plot')
a.ylabel('jumlah')
a.show()
#buat grafik stacked barplot. isinya data gender dan provinsi
```
