```python
import matplotlib  #variabel
import matplotlib.pyplot as a
import numpy as b
```

```python
x = b.arange(0.0, 2.0, 0.01) #membuat data set daei 0,0 sampai kurang dari 2,0 dengan jarak angka 0,01
x
```

```python
s = b.sin(2 * b.pi * x) #membuat data set lain berdasarkan hasil sinus dari data diatas dan dikali 2 dan dikali lagi dengan pi
s
```

```python
fig, ax = a.subplots()   #membuat grafik dengan menggunakan 00 type dengan data yang sudah dicantum diatas

ax.plot(x, s)
ax.set(xlabel='nilai x',
       ylabel='nilai sin(x)',
       title ='Visualisasi nilai sin')
ax.grid()
a.show()
```

```python
x1 = b.linspace(0.0, 5.0, 100) #membuat data set x1 dengan tehnik linspace dari 0.0 sampai 5 dengan jarak 100
x2 = b.linspace(0.0, 2.0, 100) #membuat data set x2 dengan tehnik linspace dari 0.0 sampai 2 dengan jarak 100
y1 = b.cos(2 * b.pi * x1) #membuat data set y1 dengan tehnik cos dengan data b dikali 2 dan dikali pi dan dikali lagi dengan data x1
y2 = b.cos(2 * b.pi * x2) #membuat data set y2 dengan tehnik cos dengan data b dikali 2 dan dikali pi dan dikali lagi dengan data x2
```

```python
x1
```

```python
#00 style
fig, (ax1, ax2) = a.subplots(2, 1) #membuat grafik tipe multiple subplots
fig.suptitle('multiple subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai cos(x1)')  #membuat grafik dengan nama Nilai cos(x1) dengan data x1 dan y1 dengan warna merah dan berbentuk bulat

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai cos(x2)') #membuat grafik dengan nama Nilai cos(x2) dengan data x2 dan y2 dengan warna hijau dan berbentuk titik-titik kecil

ax2.set_xlabel('Nilai x')
a.show()
```

```python
#pyplot style                     #sama kaya diatas cuman lbh ribet        
a.subplot (2, 1, 1)                  #membuat tipe grafik multiple subplots dengan data yang sudah diterapkan
a.plot (x1, y1, 'ro-')
a.title ('multiple subplots')
a.ylabel ('Nilai $cos (x_1)$')

a.subplot (2, 1, 2)             
a.plot (x2, y2, 'g.-')
a.ylabel('Nilai $cos (x_2)$')
a.xlabel('Nilai $x$')

a.show()
```

```python
mu, sigma = 100, 15 #mean, standart deviasion
x = mu + sigma * b.random.randn(10000) #normal distribution
x
```

```python
x.shape #mengecek shape
```

```python
#pyplot style
a.hist(x,
      bins = 50,
      facecolor = 'g',              #membuat grafik histogram dengan warna hijau dengan data yang diminta
      alpha = 0.75)

a.xlabel('Sumbu x')
a.ylabel('Sumbu y')
a.title ('Contoh histogram')

a.text(45, 500, r'$\mu = 100,\ \sigma=15$')
a.grid()

a.show()
```

```python
#00 style                                   #sama aja cuman beda cara dan beda warna
fig, ax = a.subplots()

ax.hist(x,
        bins = 50,
        facecolor = 'r',
        alpha = 0.75)

ax.set_xlabel('sumbu x')
ax.set_ylabel('sumbu y')
ax.set_title ('contoh histogram')

ax.text(45, 500, r'$\mu = 100,\ \sigma = 15$')
ax.grid

a.show()
```
