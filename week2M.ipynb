```python
import matplotlib                  #variabel
import matplotlib.pyplot as kucing
import numpy as tikus
```

```python
kucing.plot([2, 5, 7, 11]) #buat grafik dgn angka 2 5 7 11
kucing.ylabel('sumbu y') #memberikan nama ke garis y
kucing.show() 
```

```python
kucing.plot([2, 3, 4, 7], [5, 4, 2, 16]) #buat data grafik 2 3 4 7 untuk sumbu x sedangkan 5 4 2 16 untuk sumbu y
kucing.show()
```

```python
kucing.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') #buat data grafik 1 2 3 4 untuk sumbu x sedangkan 1 4 9 16 untuk sumbu y dengan dot 0 warna merah r
kucing.axis([0, 6, 0, 20]) # menentukan angka untuk digrafik dengan 0 - 6 untuk sumbu x dan 0 - 20 untuk sumbu y
kucing.show()
```

```python
t = tikus.arange(0., 5., 0.2) #membuat data dari anka 0,.... sampat 5,.... dengan penjumlahan 0,2
t
```

```python
kucing.plot(t, t, 'r--') #membuat data pertama dengan t dan t sebagai nilai x dan y nya dan menggunakan tipe data grafik berupa garis garis berwarna merah
kucing.plot(t, t**2, 'bs') #membuat data kedua dengan t dan t kuadrat sebagai nilai x dan y nya menggunakan tipe grafik dot kotak kotak s berwarna biru b
kucing.plot(t, t**3, 'g^') #membuat data ketiga dengan t dan t kubik sebagai nilai x dan y nya menggunakan tipe grafik dot segitiga segitiga ^ berwarna hijau g
kucing.show()
```

```python
kucing.plot(t, t, 'r--',   #sama aja cuma beda cara
            t, t**2, 'bs',
            t, t**3, 'g^')
kucing.show()
```

```python
data = {'a': tikus.arange(50),                 #membuat 3 buah susunan data dengan nama a c d
        'c': tikus.random.randint(0, 50, 50),
        'd': tikus.random.randn(50)}
data['b'] = data['a'] + 10 * tikus.random.randn(50) #membuat susunan data baru dengan nama b yang dicopy dari data a ditambah 10 dan dikali tikus.random.randn(50)
data['d'] = tikus.abs(data['d']) * 100 #membuat susunan data baru dengan mencopy dari data tikus.abs data d dikali 100
data
```

```python
kucing.scatter('a', 'b', data==data) #perintah untuk menunjukan susunan data a dan b saja 
kucing.xlabel('entry a') #menamai sumbu x dengan nama entry a
kucing.ylabel('entry b') #menamai sumbu y dengan nama entry b
kucing.show() 
```

```python
kucing.scatter('a', 'b', c='c', s='d', data=data) #perintah untuk menunjukan susunan data a dan b dengan c dibuat dengan plot color dan d dibuat sebagai sizenya
kucing.xlabel('entry a') #menamai sumbu x dengan nama entry a
kucing.ylabel('entry b') #menamai sumbu y dengan nama entry b
kucing.show() 
```

```python
names = ['group a', 'group b', 'group c'] #membuat tipe data names dengan nama group a, group b dan group c
values = [1, 10, 100] # membuat tipe data values dengan nama 1, 10, 100
```

```python
kucing.plot(names, values) #membuat grafik plot dengan data diatas
kucing.show
```

```python
kucing.scatter(names, values) #membuat grafik scatter dengan data diatas
kucing.show
```

```python
kucing.bar(names, values) #membuat grafik bar dengan data diatas
kucing.show
```
