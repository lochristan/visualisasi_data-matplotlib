```python
import matplotlib
import matplotlib.pyplot as a
import numpy as b
#variabel
```

```python
b.random.seed(2)
data = b.random.normal(loc = 100, scale = 10, size = 200)
data
#membuat dataset
```

```python
a.boxplot(data, labels = ['Data'])
a.title('Simple Box Plot')
a.xlabel('Sumbu X')
a.ylabel('Sumbu Y')
a.grid()
a.show()
#membuat box plot dengan data yang ada di atas
```

```python
a.boxplot(data, labels = ['Data'], notch = True)
a.title('Notched Box Plot')
a.xlabel('sumbu x')
a.ylabel('sumbu y')
a.show()
#mengubah box plot tipe notch dari grafik boxplot diatas
```

```python
a.boxplot(data,
          labels = ['Data'],
          showfliers = True,
          flierprops = {'markerfacecolor' : 'r', 'marker' : 'd'})
a.title('pengaturan marker')
a.xlabel('sumbu x')
a.ylabel('sumbu y')
a.show()
#perintah untuk mengatur marker untuk outlier boxplot
```

```python
a.boxplot(data, labels = ['Data'], vert = False)
a.title('horizontal box plot')
a.xlabel('sumbu x')
a.ylabel('sumbu y')
a.show()
#mengubah boxplot diatas menjadi horizontal boxplot
```

```python
b.random.seed(10)
data1 = b.random.normal(100, 10, 200)
data2 = b.random.normal(80, 30, 200)
data3 = b.random.normal(90, 20, 200)
data4 = b.random.normal(70, 25, 200)
data = [data1, data2, data3, data4]
labels = ['data 1', 'data 2', 'data 3', 'data 4']
#membuat 4 data untuk multiple boxplot
```

```python
b.random.seed(2)
data = b.random.normal(loc = 100, scale = 10, size = 200)
data
#membuat dataset
```

```python
a.violinplot(data)
a.title('simple violin plot')
a.xlabel('sumbu x')
a.ylabel('sumbu y')
a.show()
#membuat violin plot dengan dataset diatas
```

```python
a.violinplot(data, vert = False)
a.title('horizontal Violin Plot')
a.xlabel('sumbu x')
a.ylabel('sumbu y')
a.show()
#mengubah violinplot diatas menjadi horizontal
```

```python
x = b.linspace(1, 10, 25)
x
#membuat dataset
```

```python
fig, ax1 = a.subplots()
ax1.plot(x, b.exp(x), 'bs-', label = 'exp')
ax1.set_xlabel('sumbu x')
ax1.set_ylabel('exp')
ax2 = ax1.twinx()
ax2.plot(x, b.log(x), 'ro-', label = 'log')
ax2.set_ylabel('log')
fig.suptitle('contoh twin axes')
fig.legend(loc = 'upper left')
a.show()
#membuat grafik dengan kasus twinaxes dengan dataset diatas
```

```python
x = b.random.randint(16, 40, size = 30)
x
#membuat dataset
```

```python
def C2F(celsius = 0):
    return (celsius * 1.8) + 32
def konversi_suhu(ax1):
    y1, y2 = ax1.get_ylim()
    ax2.set_ylim(C2F(y1), C2F(y2))
    ax2.figure.canvas.draw()
fig, ax1 = a.subplots()
ax2 = ax1.twinx()
ax1.callbacks.connect('ylim_changed', konversi_suhu)
ax1.plot(x)
ax1.set_xlabel('Hari')
ax1.set_ylabel('Celsius')
ax2.set_ylabel('Fahrenheit')
fig.suptitle('Temperatur Udara')
a.show()
#membuat grafik dengan kasus twinaxes dengan dataset diatas
```
