```python
import matplotlib
import matplotlib.pyplot as a
import numpy as b
#variabel
```

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
#buat data dan kategori
```

```python
a.bar(kategori, data)
a.grid()
a.xlabel('kategori')
a.ylabel('jumlah')
a.title ('contoh bar plot')
a.show()
#membuat grafik jenis bar dengan data diatas
```

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
#buat data dan kategori baru
```

```python
a.bar(kategori,
      data,
      color='red',
      alpha=0.25) #ketebalan warna
a.grid(linestyle='--',
       linewidth=2,
       axis='x',
       alpha=0.50) #ketebalan garis putus"
a.xlabel('kategori')
a.ylabel('jumlah')
a.title('contoh bar plot')
a.show()
#buat grafik bar dengan warna merah dan grid garis garis putus dari garis x
```

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
#buat data baru 2 biji dan kategori baru
```

```python
a.bar(kategori, data1, label='Data 1')
#a.bar(kategori, data2, label='Data 2')
a.bar(kategori, data2, label='Data 2', bottom=data1)
a.legend()
a.grid(linestyle='--',
       linewidth=1,
       axis='y',
       alpha=0.75)
a.xlabel('kategori')
a.ylabel('kategori')
a.title('contoh stacked bar plot')
a.show()
#perintah untuk membuat stacked bar plot. Pengertiannya data1 dan data2 digabung. jadi data1 dlo yang ditaro ssdh itu ditambah data2
#karena bottom=data1 di baris code 3
```

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
x=b.arange(len(kategori))
x
#buat data
```

```python
width = 0.35
a.bar(x-width/2, data1, width, label='Data1')
a.bar(x+width/2, data2, width, label='Data2')
a.xticks(x, kategori)
a.grid(linestyle='--',
       linewidth=1,
       axis='y',
       alpha=0.75)
a.xlabel('kategori')
a.ylabel('jumlah')
a.title('contoh grouped abr plot')
a.legend()
a.show()
#perintah untuk membuat grouped bar plot
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
#buat kategori dan data
```

```python
a.pie(data,
      labels=kategori,
      autopct='%1.1f%%',
      startangle=90)
a.title('contooh pie plot')
a.show()
#perintah buat grafik bulet dengan data diatas
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
#buat kategori dan data dan warna
```

```python
a.pie(data,
      labels=kategori,
      colors=warna,
      autopct='%1.1f%%',
      startangle=90)
a.legend()
a.title('contoh colored pie plot')
a.show()
#buat grafik bulet dengan data diatas dan diwarna dengan warna yang diinginkan
```

```python
a.pie(data,
      labels=kategori,
      colors=warna,
      autopct='%1.1f%%',
      startangle=90)
a.legend()
a.title('contoh colored pie plot')
a.show()
#buat grafik bulet dengan data diatas dan diwarna dengan warna yang diinginkan
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08]
#buat kategori dan data dan warna
```

```python
a.pie(data,
      labels=kategori,
      colors=warna,
      explode=explode_var,
      autopct='%1.1f%%',
      startangle=90)
a.title('contoh exploded pie plot')
a.show()
#membuat grafik bulet yang dipecah berdasarkan data diatas
```

```python
data_uas=[
    ['Bejo', 70],
    ['Tejo', 83],
    ['Cecep', 62],
    ['Wati', 74],
    ['Karti', 71]
]
#membuat pendataan
```

```python
data_uas #memunculkan data
```

```python
fig, ax = a.subplots()
table = a.table(cellText=data_uas,
               loc='center')
table.set_fontsize(14)
table.scale(1, 4) #ukuran kolom dan baris
ax.axis(False) #kalo diiliangin bag ini bakal nongol tabel buat grafik 
a.show() 
#buat tabel 00style dengan data diatas
```

```python
table = a.table(cellText=data_uas,
                loc='center')
table.set_fontsize(14)
table.scale(1, 4)
ax = a.gca()
ax.axis(False)
a.show()
#sama kaya diatas cuman inipyplot style
```
