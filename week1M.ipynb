```python
import matplotlib.pyplot as lo #variabel
import numpy as uis
```

```python
x = [1, 2, 3, 4] #data
y = [1, 4, 2, 3]
lo.plot(x, y) #perintah untuk menampilkan grafik berdasarkan data diatas
```

```python
fig = lo.figure() #perintah untuk membuat figure tanpa tabel grafik
```

```python
fig, ax = lo.subplots() #perintah untuk membuat figure dengan tabel grafik
```

```python
fig, axs = lo.subplots(2, 3)#perintah untuk membuat figure dengan tabel grafik 2 x 3
```

```python
x = uis.linspace(1, 7, 100) #membuat sejumlah bilangan dengan 3 parameter 1, 7, 100
x         #catetan:D. 1 disini tu starting point 7 nilai akhir 100 itu total data yang bakal kebentuk mulai dari 1 - 7
                       #ini oo style
```

```python
fig, ax = lo.subplots() #pembuatan figure dan tabel grafik

ax.plot(x, x, label = 'linear')         #plot 3 variant ke tabel frafik
ax.plot(x, x**2, label = 'quadratic')
ax.plot(x, x**3, label = 'cubic')

ax.set_xlabel('x label') #membuat label x ke tabel grafik
ax.set_ylabel('y label') #membuat label y ke tabel grafik
ax.set_title("simple plot") # membuat judul tabel grafik
ax.legend() #membuat legenda
```

```python
x = uis.linspace(1, 5, 100) #pengertiannya sama seperti yang diatas
x                             #pyplot style
```

```python
lo.plot(x, x, label = 'linear')       #plot 3 variant ke tabel frafik
lo.plot(x, x**2, label = 'quadratic')
lo.plot(x, x**3, label = 'cubic')

lo.xlabel('x label') #membuat label x ke tabel grafik
lo.ylabel('y label')  #membuat label y ke tabel grafik
lo.title("simple plot") # membuat judul tabel grafik
lo.legend() #membuat legenda
```
