```python
import matplotlib
import matplotlib.pyplot as a
import numpy as b
#variabel
```

```python
a.style.available
#menampilkan setiap style yg ada
```

```python
a.style.use('ggplot')
```

```python
x = b.linspace(1, 10, 15)
a.plot(x, x, 'bs-', label = 'lin')
a.plot(x, b.log(x), 'ro-', label = 'log')
a.legend()
a.xlabel('kategori')
a.ylabel('jumlah')
a.title('contoh bar plot')
a.show()
#buat grafik
```

```python
a.style.use(['seaborn', 'dark_background'])
```
