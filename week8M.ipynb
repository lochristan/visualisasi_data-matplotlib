```python
import matplotlib
import matplotlib.pyplot as a
import numpy as b
```

```python
data = {'item a' : 109438,
        'item b' : 103569,
        'item c' : 112214,
        'item d' : 112591,
        'item e' : 100934,
        'item f' : 103660,
        'item g' : 137351,
        'item h' : 123381,
        'item i' : 135841,
        'item j' : 104437}
data
#buat data
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

```python
fig, ax = a.subplots()
ax.barh(items, count)
a.show()
#buat grafik dengan data diatas
```

```python
a.style.available
#memperlihatkan style yg tersedia
```

```python
a.style.use('bmh')
fig, ax = a.subplots()
ax.barh(items, count)
a.show()
#menambahkan style di gradik diatas
```

```python
fig, ax = a.subplots()
ax.barh(items,count)
labels = ax.get_xticklabels()
a.setp(labels,
       rotation = 45,
       horizontalalignment = 'right')
a.show()
#mengatur tick table
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan (x, pos):
    return f'{int(x/1000)}K'
```

```python
fig, ax = a.subplots()
ax.barh(items,count)
formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)
ax.set(title = 'Contoh Plot',
       xlabel = 'Jumlah',
       ylabel = 'items')
a.show()
#pengaturan format tick tabel
```

```python
fig, ax = a.subplots()
ax.barh(items, count)
ax.set(title = 'Contoh Plot',
       xlabel = 'jumlah',
       ylabel = 'Items')
a.show()
#pengaturan label
```

```python
fig, ax = a.subplots()
ax.barh(items, count)
ax.axvline(b.mean(count),
           ls = '--',
           color = 'r')
ax.set(title = 'Contoh Plot',
       xlabel = 'Jumlah',
       ylabel = 'Items')
a.show()
#penambahan garis
```
